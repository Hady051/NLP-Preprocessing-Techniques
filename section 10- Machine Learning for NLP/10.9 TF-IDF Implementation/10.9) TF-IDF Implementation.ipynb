```python
import pandas as pd
```

```python
messages = pd.read_csv('smsspamcollection.csv')
```

```python
messages
```

### <span style="color:orange">Data Cleaning and Preprocessing</span>

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
word_net_lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
```

```python
for i in range(0, len(messages) ):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i] )
    # review = review.lower()
    review = review.split()
    review = [word_net_lemmatizer.lemmatize(word, pos='v') for word in review
              if word not in set(stopwords.words('english') )]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## <span style="color:orange">TF-IDF and N-grams</span>

### <span style="color:orange">TF-IDF alone.</span>

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf_idf = TfidfVectorizer(max_features=100)

independent_features = tf_idf.fit_transform(corpus).toarray()
```

### <span style="color:orange">TF-IDF with N-grams</span>

```python
tf_idf_ngram = TfidfVectorizer(max_features=100, ngram_range=(2, 3) )
```

```python
corpus2 = []
```

```python
for i in range(0, len(messages) ):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i] )
    # review = review.lower()
    review = review.split()
    review = [word_net_lemmatizer.lemmatize(word, pos='v') for word in review
              if word not in set(stopwords.words('english') )]
    review = ' '.join(review)
    corpus2.append(review)
```

```python
corpus2
```

```python
independent_features2 = tf_idf_ngram.fit_transform(corpus2).toarray()
```

```python
print(independent_features2)
```
