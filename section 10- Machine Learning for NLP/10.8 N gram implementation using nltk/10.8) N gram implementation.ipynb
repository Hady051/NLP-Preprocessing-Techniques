#### <span style="color:orange">Data Cleaning and Preprocessing.</span>

```python
import re
from nltk.corpus import (stopwords)
from nltk.stem import PorterStemmer
```

##### Lemmitization

```python
from nltk.stem import WordNetLemmatizer
```

```python
word_net_lemmatizer = WordNetLemmatizer()
```

```python
corpus = []

for i in range(0, len(messages) ):
    review = re.sub('[^a-z A-Z]', ' ', messages['message'][i] )
    review = review.lower()
    review = review.split()
    review = [word_net_lemmatizer.lemmatize(word, pos='v') for word in review
              if not word in set(stopwords.words('english') ) ]
    review = ' '.join(review)
    corpus.append(review)
```

### <span style="color:orange">Create Bag Of Words (BOG) model.</span>

```python
## Create the Bag OF Words model
from sklearn.feature_extraction.text import CountVectorizer
```

```python
## for Binary BOW enable binary = True
count_vectorizer = CountVectorizer(max_features=500)
```

```python
input_features = count_vectorizer.fit_transform(corpus).toarray()
```
